```python
# Data manipulation and analysis
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Date handling
from datetime import datetime

# Display options
pd.set_option('display.max_columns', 100)
sns.set(style="whitegrid", palette="pastel")

# Load datasets
client_df = pd.read_csv("client_data (1).csv", parse_dates=['date_activ', 'date_end', 'date_modif_prod', 'date_renewal'])
price_df = pd.read_csv("price_data (1).csv", parse_dates=['price_date'])

######## Preview the datasets
print("Client Data:")
display(client_df.head(2))

print("Price Data:")
display(price_df.head(2))

######### Check for missing values
print("Missing values in Client Data:")
display(client_df.isnull().sum())

print("\nMissing values in Price Data:")
display(price_df.isnull().sum())

############### Check data types
print("Data types in Client Data:")
display(client_df.dtypes)

print("\nData types in Price Data:")
display(price_df.dtypes)

########## Descriptive statistics for numeric columns
print("Client Data - Numerical Descriptive Stats:")
display(client_df.describe())

print("\nPrice Data - Numerical Descriptive Stats:")
display(price_df.describe())

############# Desc.stat. for categorical variables
print("\nClient Data - Categorical Value Counts:")
categorical_cols = client_df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\n{col}:\n", client_df[col].value_counts().head())

############## Distribution of churn variable
sns.countplot(data=client_df, x='churn')
plt.title('Churn Distribution')
plt.xlabel('Churned')
plt.ylabel('Number of Clients')
plt.show()

print(client_df['churn'].value_counts(normalize=True))

###################### Numerical distribution plots
num_cols = ['cons_12m', 'cons_gas_12m', 'cons_last_month', 
            'forecast_cons_12m', 'forecast_cons_year',
            'net_margin', 'margin_net_pow_ele', 'margin_gross_pow_ele',
            'forecast_discount_energy', 'forecast_meter_rent_12m',
            'forecast_price_energy_off_peak', 'forecast_price_energy_peak',
            'forecast_price_pow_off_peak', 'nb_prod_act', 'pow_max']

client_df[num_cols].hist(bins=30, figsize=(20, 15), edgecolor='black')
plt.suptitle('Histograms of Key Numeric Features', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()

# Boxplots to compare net margin with churn
for col in ['margin_net_pow_ele']:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=client_df, x='churn', y=col)
    plt.title(f'{col} Distribution by Churn')
    plt.show()

# Categorical distributions vs churn
for col in ['channel_sales', 'origin_up']:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=client_df, x=col, hue='churn')
    plt.title(f'{col} vs Churn')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
